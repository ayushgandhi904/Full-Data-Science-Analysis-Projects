## Outlier effect

```python
# If we want clean data, we need to know effect of outlier & how does it effect 
```

```python
plt.figure(figsize = (18, 18))

for i, feature in enumerate(na_features, 1):
    plt.subplot(5, 5, i)
    data = train.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    f = data.groupby(feature)["SalePrice"].median()
    _ = sns.barplot(x = f.index, y = f.values)
    
    plt.title(feature)
    plt.xticks([0, 1], ["Present", "Missing"])
    plt.ylabel("Sales Price")
plt.show()
    
    
```

### Normality

```python
#distplot & QQ plot

sns.distplot(train["SalePrice"], fit = norm);
(mu, sigma) = norm.fit(train["SalePrice"])
print(f"mu = {mu}  & sigma = {sigma}")

plt.legend([f"normal distribution {mu:.2f} & sigma {sigma : .2f}"], loc = "best")
plt.ylabel("Frequency")
plt.title("SalePrice distribution")

fig = plt.figure()
res = stats.probplot(train["SalePrice"], plot = plt)
plt.show()
```

```python
#logarithmic transformation 

data = train.copy()
sale_price = np.log(train["SalePrice"])

for i, feature in enumerate(continuous_features[:-1], 1):
    data = train[feature].copy()
    
    if 0 in data.unique():
        continue
    else: 
        data = np.log(data)
        data.name = feature
        _ = plt.figure(figsize = (15, 8))
        plt.subplot(1, 2, 1)
        sns.regplot(data, sale_price, fit_reg = True, scatter_kws = {"color":"blue"}, line_kws={"color":"red"}).set_title(f"Correlation: {data.corr(sale_price)}")
        plt.subplot(1, 2, 2)
        sns.distplot(data).set_title(f"log of {feature}")
    plt.show()
```
